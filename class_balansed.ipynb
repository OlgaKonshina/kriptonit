```python
import pandas as pd
```

```python
df = pd.read_csv('/content/train.csv')
```

```python
df
```

```python
disgust_df = df[df.disgust == 1]
disgust_df = pd.concat([disgust_df for i in range(8)])
disgust_df
```

```python
anger_df = df[df.anger == 1]
anger_df = pd.concat([anger_df for i in range(2)])
anger_df
```

```python
fear_df = df[df.fear == 1]
fear_df = pd.concat([fear_df for i in range(10)])
fear_df
```

```python
joy_df = df[df.joy == 1]
joy_df = joy_df.sample(frac=0.7)
joy_df
```

```python
sadness_df = df[df.sadness == 1]
sadness_df = pd.concat([sadness_df for i in range(2)])
sadness_df
```

```python
surprise_df = df[df.surprise == 1]
surprise_df = pd.concat([surprise_df, surprise_df])
surprise_df
```

```python
neutral_df = df[df.neutral	 == 1]
neutral_df = neutral_df.sample(frac=0.8)
neutral_df
```

```python
labels = ["anger", "disgust", "fear", "joy", "sadness", "surprise", "neutral"]
```

```python
df_train = pd.concat([anger_df, disgust_df, fear_df, joy_df, sadness_df, surprise_df, neutral_df])
```

```python
df_train = df_train.sample(frac=1).reset_index().drop(columns=['index'])
```
